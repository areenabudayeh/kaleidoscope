### Exploratory Data Analysis (EDA)

```python
# rows and columns
data.shape
```

```python
# Features
data.columns
```

```python
data.info()
```

```python
# No missing values
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap

# Visualize missing values using a heatmap

custom_cmap = LinearSegmentedColormap.from_list("custom_pistachio", [ "#95bfb2","white"])

plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cmap=custom_cmap, cbar=False)
plt.title('Missing Values Heatmap', fontsize=14)
plt.show()
```

```python
# Find duplicate rows
duplicates = data[data.duplicated()]
print(f"Number of duplicate rows: {duplicates.shape[0]}")
```

```python
# Check for unique values in categorical columns to see if any unexpected categories exist
for column in data.select_dtypes(include=['object']).columns:
    print(f'Unique values in {column}:')
    print(data[column].value_counts())
    print('\n')
```

```python
for i in range(1, 9):
    episode_column = f'Episode {i}'
    print(data[episode_column].value_counts)
```

```python
for i in range(1, 9):
    time_period_column = f'Time period Episode {i}'
    print(data[time_period_column].value_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Bar chart
data['max_label'].value_counts().plot(kind='bar', color='#95bfb2')
plt.title('max_label Counts')
plt.xlabel('max_label')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns

custom_cmap = LinearSegmentedColormap.from_list("custom_pistachio", [ "#95bfb2","white"])

crosstab = pd.crosstab(data['Genres Episode 1'], data['max_label'])
sns.heatmap(crosstab, annot=True, cmap=custom_cmap, fmt='d')
plt.title('Heatmap of Genres Episode 1 vs max_label')
plt.xlabel('max_label')
plt.ylabel('Genres Episode 1')
plt.show()
```

```python
# Histogram for all numerical features

data.hist(figsize=(12, 10), bins=20, color='#95bfb2', edgecolor='black')
plt.tight_layout()
plt.show()
```
